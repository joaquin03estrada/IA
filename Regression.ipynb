```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Ejemplos:

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Feature (independent variable)
y = np.array([2, 4, 5, 4, 5])  # Target (dependent variable)
```

Se crea el modelo de regresion lineal

```python
model = LinearRegression()
```

Ajustamos el modelo a los datos

```python
model.fit(X, y)
```

Predecimos usando el modelo

```python
y_pred = model.predict(X)
```

Graficamos, añadimos etiquetas y titulo

```python
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
